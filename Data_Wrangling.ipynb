```python
import pandas as pd
import zipfile
import os
```

```python
# seeing how the folders are structured in ZIP folder

#copied from https://stackoverflow.com/questions/6510477/how-can-i-list-only-the-folders-in-zip-archive-in-python
# from Dave B. answer
z = zipfile.ZipFile('archive.zip', 'r')
dirs = list(set([os.path.dirname(x) for x in z.namelist()]))
print(dirs)
```

```python

for file in z.namelist():
    z.extract(file,'data')
```
